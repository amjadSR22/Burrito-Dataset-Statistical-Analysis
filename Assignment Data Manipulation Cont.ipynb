```python
burritos_data['overall'].fillna(value=burritos_data['overall'].mean(), inplace=True)
```

```python
print(burritos_data['overall'])
```

```python
burritos_data['Hunger'].fillna(value=burritos_data['Hunger'].median(), inplace=True)
```

```python
print(burritos_data['Hunger'])
```

```python
burritos_data['Synergy'].fillna(burritos_data['Synergy'].mode(), inplace=True)
```

```python
print(burritos_data['Synergy'])
```

```python
burritos_data['Cheese'] = burritos_data['Cheese'].fillna(8)
```

```python
print(burritos_data['Cheese'])
```

```python
burritos_data['Fries'].fillna(method="ffill")
```

```python
burritos_data['Guac'].fillna(method="bfill")
```

```python
burritos_data.drop(columns=['URL', 'Mass (g)','Density (g/mL)','Circum','Volume','Rec','Reviewer','Notes','Unreliable','NonSD','Beef','Pico','Guac','Cheese','Fries','Sour cream','Pork','Chicken','Shrimp','Fish','Rice','Beans'
                           ,'Lettuce','Tomato','Bell peper','Carrots','Cabbage','Sauce','Salsa.1','Cilantro','Onion','Taquito','Pineapple',
                           'Ham','Chile relleno','Nopales','Lobster','Queso','Egg','Mushroom','Bacon','Sushi','Avocado','Corn','Zucchini'])
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=burritos_data['Hunger'], y=burritos_data['Neighborhood'])
plt.xlabel("Hunger levels")
plt.ylabel("Neighborhood")
plt.title("Hunger levels based on neighborhoods")
plt.show()
```

```python
plt.figure(figsize=(10,5));
ax = plt.subplot()
ax.bar(burritos_data["Google"], burritos_data["Cost"])
ax.bar(burritos_data["Yelp"], burritos_data["Cost"], color='yellow')
plt.title("Difference in rating on Google and Yelp")
```

```python
plt.figure(figsize=(10,5));
sns.barplot(x=burritos_data['Yelp'], y=burritos_data['Cost'])
plt.title("Relationship between cost and Yelp rating")
plt.ylabel("Cost")
plt.xlabel("Yelp ratings")
```

```python
#Histogram chart
plt.figure(figsize=(20,8))
sns.histplot(data=burritos_data, x='Hunger', hue='Google')
plt.title("Hunger levels of customers after eating a burrito based on their Google rating")
plt.ylabel("Hunger levels")
plt.xlabel("Google ratings")
```

```python
plt.figure(figsize=(10,5));
plt.scatter(burritos_data['Google'], burritos_data['Cost']);
plt.title("Relationship between cost and Google rating")
plt.ylabel("Cost")
plt.xlabel("Google ratings")
```

```python
burritos_data[['Yelp','Google']].plot.box();
plt.title("Difference in rating on Google and Yelp")
plt.ylabel("Rating")
plt.xlabel("Platform")
```

```python
sns.kdeplot(data=burritos_data['Tortilla'],shade=True)
plt.title("The amount of orders for Tortilla")
plt.ylabel("Orders")
plt.xlabel("Tortilla")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.kdeplot(data=burritos_data['Google'], shade=True)
sns.kdeplot(data=burritos_data['Yelp'], shade=True)
plt.title("Ratings in Google and Yelp")
plt.xlabel("Ratings")
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5));
sns.distplot(x=burritos_data['Cost']);
plt.title("The cost of orders for Tortilla")
plt.xlabel("Cost")
```

```python
plt.figure(figsize=(10,5));
rating = ['Platforms']
Google = burritos_data['Google']
Yelp = burritos_data['Yelp']
x_axis = np.arange(len(rating))
plt.bar(x_axis -0.2, Google, width=0.4, label = 'Google')
plt.bar(x_axis +0.2, Yelp, width=0.4, label = 'Yelp')
plt.xticks(x_axis, rating)
plt.legend()
plt.ylabel("Rating")
plt.show()
```

```python
num_yelp = burritos_data['Yelp'].sum()
num_google = burritos_data['Google'].sum()
num_rating = [num_yelp,num_google]
label = ['Yelp' , 'Google']
sns.color_palette('pastel')
plt.figure(figsize=(10,20))
plt.pie(num_rating, labels = label)
plt.legend(loc = 4 );
```
